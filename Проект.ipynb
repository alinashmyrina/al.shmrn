```python
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python
link = "https://www.hse.ru/ba/political/tutors" 
```

```python
page = requests.get(link)
soup = BeautifulSoup(page.text, 'lxml')
```

```python
tutors = []
for i in soup.findAll('a'):
    if '/mirror/' in i['href']:
        tutors.append(i['href']) 
```

```python
tutors 
```

```python
full_tutors = []
for i in tutors:
    res = 'https://www.hse.ru/' + i
    full_tutors.append(res) 
```

```python
full_tutors
```

```python
physical_links = []
for link in full_tutors:
    page1 = requests.get(link)
    soup1 = BeautifulSoup(page1.text)
    res1 = list(soup1.find_all('a', {'href': '/org/hse/ouk/fizvos/'}))
    if res1 == []:
        continue
    physical_links.append(link)
    print(link + '.....done') 
    del page1
    del soup1
```

```python
new_tutors = []
for item in set(full_tutors).difference(physical_links):
    new_tutors.append(item)
```

```python
new_tutors
```

```python
english_links = []
for link in new_tutors:
    page1 = requests.get(link)
    soup1 = BeautifulSoup(page1.text)
    res = list(soup1.find_all('a', {'href': 'https://lang.hse.ru/'}))
    if res == []:
        print('       ')
        continue
    english_links.append(link)
    print(link + '.....done') 
    del page1
    del soup1
```

```python
new_tutors1 = []
for item in set(new_tutors).difference(english_links):
    new_tutors1.append(item)
```

```python
url0 = new_tutors1[0]
page0 = requests.get(url0)
soup0 = BeautifulSoup(page0.text) 
```

```python
name = soup0.findAll('meta', {'property': 'og:title'})[0]['content']
name 
```

```python
interest = soup0.findAll('a', {'class':'tag tag_small'})
interest 
```

```python
interests = []
for i in interest:
    res = list(i)
    interests.append(res)
interests 
```

```python
for i in interests:
    i.append(name)
interests 
```

```python
def GetInterests(url0):
    """
    Returns a dictionary with professional interest as key and names of tutors as values
    Parameters:
    
    url0 is a link to the personal page of tutor on the official site of Higher School of Economics
    """
    page0 = requests.get(url0)
    soup0 = BeautifulSoup(page0.text)
    name = soup0.findAll('meta', {'property': 'og:title'})[0]['content']
    interest = soup0.findAll('a', {'class':'tag tag_small'})
    interests = []
    for i in interest:
        res = list(i)
        interests.append(res)
    for i in interests:
        i.append(name)
    return dict(interests)
```

```python
info = []
for link in new_tutors1:
    res = GetInterests(link)
    info.append(res)
    sleep(2)
    print(link+'.......done')
```
